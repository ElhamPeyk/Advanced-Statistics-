```python
# Modeling the number of meteorites that fall into the ocean in a year
```

```python
z1=3 # means it is Geometric Distribution
z2=0.98 # means The probability of success(P)
```

```python
print(f"z1 = {z1}, z2 = {z2}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
p= z2
#Possible values of K ​​(until the probability is less than 0.5%)

k_values = np.arange(1, 15)
probs = (1 - p) ** (k_values - 1) * p #Geometric Distribution

# probability diagram
plt.figure(figsize=(8, 5))
plt.bar(k_values, probs, color='blue', alpha=0.6, label='P(X=k)')
plt.xlabel('Number of tests(k)')
plt.ylabel('ProbabilityP(z=k)')
plt.title('Geometric Distribution(p=0.98)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
expected_value = 1 / p

median_value = np.ceil(np.log(0.5) / np.log(1 - p))

print(f"E[X]: {expected_value:.2f}")
print(f"median: {median_value}")
```
