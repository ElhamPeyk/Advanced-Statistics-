```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
```

Uptime data before failure (hours)

```python
data = np.array([167, 1, 93, 46, 64])
```

theta is the mean of the data:

```python
theta_mle = np.mean(data)
```

PDF:

```python
x = np.linspace(0, 200, 1000)
pdf_values = expon.pdf(x, scale=theta_mle)
```

PDF Diagram:

```python
plt.figure(figsize=(8, 5))
plt.plot(x, pdf_values, label=f'Exponential PDF (Î¸ = {theta_mle:.2f})', color='blue')
plt.xlabel("Operating time before failure (hours)")
plt.ylabel("Probability densityf(T)")
plt.title("Probability density function (PDF) for router uptime")
plt.legend()
plt.grid()
plt.show()
```

Probability that a router will fail before 50 hours:

```python
p_less_50 = expon.cdf(50, scale=theta_mle)
print(f"Probability that a router will fail before 50 hours: {p_less_50:.4f}")
```

Mean and variance:

```python
expected_value = theta_mle  # In an exponential distribution, the mean is equal to Î¸.
variance = theta_mle ** 2  # In an exponential distribution, the variance is equal to Î¸^2.
```

```python
print(f"Average router uptime: {expected_value:.2f} hour")
print(f"Router uptime variance: {variance:.2f}")
```
